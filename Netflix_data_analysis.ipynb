```python
import pandas as pd

# Load datasets
low = pd.read_csv("15 Lowest-Grossing Movies Of All Time.csv")
high = pd.read_csv("Top 1000 Highest-Grossing Movies Of All Time.csv")

print("=== Highest Grossing ===")
print(high.head())

print("\n=== Lowest Grossing ===")
print(low.head())
```

```python
low['category'] = 'lowest'
high['category'] = 'highest'
```

```python
movies = pd.concat([low, high], ignore_index=True)
```

```python
movies['IMDb Rating'] = pd.to_numeric(movies['IMDb Rating'], errors='coerce')
movies['Runtime (mins)'] = pd.to_numeric(movies['Runtime (mins)'], errors='coerce')
movies['Num Votes'] = pd.to_numeric(movies['Num Votes'], errors='coerce')
```

```python
movies['Genres'] = movies['Genres'].str.split(', ')
```

```python
print("✅ First 5 rows of combined dataset:")
print(movies.head())
print("\n✅ Missing values in each column:")
print(movies.isnull().sum())
```

```python
# Average IMDb rating for each category
avg_rating = movies.groupby('category')['IMDb Rating'].mean()
print("Average IMDb Rating:")
print(avg_rating)
```

```python
# Average runtime for each category
avg_runtime = movies.groupby('category')['Runtime (mins)'].mean()
print("\nAverage Runtime (mins):")
print(avg_runtime)
```

```python
from collections import Counter

# Split genres lists into single list
high_genres = movies[movies['category']=='highest']['Genres'].sum()
low_genres = movies[movies['category']=='lowest']['Genres'].sum()

print("\nTop Genres in Highest Grossing Movies:")
print(Counter(high_genres).most_common(5))

print("\nTop Genres in Lowest Grossing Movies:")
print(Counter(low_genres).most_common(5))
```

```python
# Save the cleaned dataset
movies.to_csv("movies_cleaned.csv", index=False)

# Optional: Prepare genres for Power BI (one genre per row)
genres_df = movies.explode('Genres')
genres_df.to_csv("movies_genres.csv", index=False)
```
